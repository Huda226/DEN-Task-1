```python
 #for conversion yes,no into 0,1
 encoder = LabelEncoder()
```

```python
enco_col = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
for col in enco_col:
    df[col] = encoder.fit_transform(df[col])
```

```python
df
```

```python
# applying model here
x=df.drop(columns=['price'],axis=1)
y=df['price']
```

```python
scalar=MinMaxScaler()
x=scalar.fit_transform(x)
y=scalar.fit_transform(y.values.reshape(-1,1))
```

```python
x

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
In_model=LinearRegression()
In_model.fit(x_train,y_train)
```

```python
y_predicted=In_model.predict(x_test)
```

```python
mean_a_e=mean_absolute_error(y_test,y_predicted)
mean_s_e=mean_squared_error((y_test),y_predicted)
```

```python
# accuracy
acc= r2_score(y_test,y_predicted)
acc
```
